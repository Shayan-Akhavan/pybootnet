# Data Importing and Pre-processing

```python
# Load the data from the files
species_data = pd.read_csv('binned_species_read_counts_GTDBTk_BLAST_clr.csv')
metadata = pd.read_csv('feces_metadata_weeks3_10_cohortB.csv')  
```

```python
species_data
```

```python
map_columns(species_data, "Species")
```

```python
# Changing the Feature ID's into X1, X2, X3...
string_map = {}
counter = 1

for value in species_data['Species'].unique():
    string_map[value] = f'X{counter}'
    counter += 1

# Map the values in the "feature.ID" column using the dictionary
species_data['Species'] = species_data['Species'].map(string_map)
```

```python
# Transposing df2 and renaming the index column
species_transposed = species_data.set_index('Species').T
species_transposed.reset_index(inplace=True)
species_transposed.rename(columns={'index': 'Mouse_Label_Cols'}, inplace=True)

# Merging df1 with the transposed df2
combined_df = pd.merge(metadata, species_transposed, on='Mouse_Label_Cols', how='left')
```

```python
combined_df = combined_df.drop(["Mouse_Label_Cols","Mouse","Weight","Mouse_Week"], axis=1)
```

```python
week3_male_wild = combined_df[(combined_df['Week'] == 3) & (combined_df['Group_Name'] == 'MaleWildType')]

week3_male_mutant = combined_df[(combined_df['Week'] == 3) & (combined_df['Group_Name'] == 'MaleMutant')]

week3_female_wild = combined_df[(combined_df['Week'] == 3) & (combined_df['Group_Name'] == 'FemaleWildType')]

week3_female_mutant = combined_df[(combined_df['Week'] == 3)  & (combined_df['Group_Name'] == 'FemaleMutant')]
```

```python
week10_male_wild = combined_df[(combined_df['Week'] == 10) & (combined_df['Group_Name'] == 'MaleWildType')]

week10_male_mutant = combined_df[(combined_df['Week'] == 10) & (combined_df['Group_Name'] == 'MaleMutant')]

week10_female_wild = combined_df[(combined_df['Week'] == 10) & (combined_df['Group_Name'] == 'FemaleWildType')]

week10_female_mutant = combined_df[(combined_df['Week'] == 10)  & (combined_df['Group_Name'] == 'FemaleMutant')]
```

```python
# Drop 'Week' column after creating the separate DataFrames
week3_male_wild = week3_male_wild.drop(['Week'], axis=1)
week3_male_mutant = week3_male_mutant.drop(['Week'], axis=1)
week3_female_wild = week3_female_wild.drop(['Week'], axis=1)
week3_female_mutant = week3_female_mutant.drop(['Week'], axis=1)
week10_male_wild = week10_male_wild.drop(['Week'], axis=1)
week10_male_mutant = week10_male_mutant.drop(['Week'], axis=1)
week10_female_wild = week10_female_wild.drop(['Week'], axis=1)
week10_female_mutant = week10_female_mutant.drop(['Week'], axis=1)
```

```python
# Drop 'Week' column after creating the separate DataFrames
week3_male_wild.to_csv('week3mw.csv', index=False)
week3_male_mutant.to_csv('week3mm.csv', index=False)
week3_female_wild.to_csv('week3fw.csv', index=False)
week3_female_mutant.to_csv('week3fm.csv', index=False)
week10_male_wild.to_csv('week10mw.csv', index=False)
week10_male_mutant.to_csv('week10mm.csv', index=False)
week10_female_wild.to_csv('week10fw.csv', index=False)
week10_female_mutant.to_csv('week10fm.csv', index=False)
```
